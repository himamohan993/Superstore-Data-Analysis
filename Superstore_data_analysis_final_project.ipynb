DATA PRE-PROCESSING

```python
# Data Pre-processing
```

```python
# Check missing values
```

```python
df.isnull().sum()
```

```python
# Handle missing values
```

```python
df['Postal Code'] = df['Postal Code'].fillna(0)
```

```python
#remove missing values
df = df.dropna()
```

```python
# Remove Duplicates
print("Duplicate rows:", df.duplicated().sum())

df = df.drop_duplicates()
```

```python
#Correcting Data type

df.dtypes
```

```python
# Convert date columns to datetime
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Create derived columns
```

```python
# Delivery time in days
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
# Profit margin
df['Profit Margin'] = df['Profit'] / df['Sales']
```

# Visualizations

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.style.use('default')
```

Sales by Category (Bar Plot)

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

Profit by Region (Bar Plot)

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Region")
plt.show()
```

Sales Distribution (Histogram)

```python

plt.hist(df['Sales'], bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

Discount vs Profit (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['Sales','Quantity','Discount','Profit']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Sales by Sub-Category (Horizontal Bar Chart)

```python
plt.figure(figsize=(10,6))
sns.barplot(y='Sub-Category', x='Sales', data=df, estimator=sum)
plt.title("Total Sales by Sub-Category")
plt.show()
```

Profit by Category





```python
profit_by_category = df.groupby('Category')['Profit'].sum()
profit_by_category
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
plt.pie(profit_by_category,
        labels=profit_by_category.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Profit by Category")
plt.show()
```

Monthly Sales Trend (Line Chart)

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

Region-wise Sales Share (Pie Chart)

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by Region")
plt.show()
```

Heatmap: Category vs Region

```python
pivot = pd.pivot_table(df, values='Profit', index='Region', columns='Category', aggfunc='sum')

plt.figure(figsize=(8,5))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title("Profit by Region and Category")
plt.show()
```

Top 10 Products by Sales

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=75)
plt.show()
```
